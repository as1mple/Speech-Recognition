# Pipeline

```python
res = []
errors = []

for user_id in tqdm(df.user_id.unique()):
  os.makedirs(f"{res_folder_path}/{user_id}") if not os.path.exists(f"{res_folder_path}/{user_id}") else "exist"

  columns = ['user_id', 'index', 'description']
  user_df = df[df['user_id'] == user_id][columns].reset_index(drop=True)
  for idx, row in user_df.iterrows():
      index = row['index']
      path_to_wav = f"{user_id}/{index}.wav"

      description = row['description']
      try:
        substrings = "\n".join(split_string_into_substrings(description))
      except Exception as e:
        errors.append(index)
        continue

      temp_text_file = tempfile.NamedTemporaryFile(delete=False, mode='w', suffix='.json')
      temp_text_file.write(substrings)
      temp_text_file.close()

      config_string = f"task_language=uk|is_text_type=plain|os_task_file_format=json"
      aeneas_task = task.Task(config_string=config_string)
      try:
        aeneas_task.audio_file_path_absolute = f"{os.getcwd()}/{path_to_wav}"
      except Exception as e:
        errors.append(index)

      else:
        aeneas_task.text_file_path_absolute = temp_text_file.name
        aeneas_task.sync_map_file_path_absolute = f"{res_folder_path}/{user_id}/{index}.json"

        executor = et.ExecuteTask(aeneas_task)
        executor.execute()

        aeneas_task.output_sync_map_file()

        os.unlink(temp_text_file.name)

        audio = AudioSegment.from_file(aeneas_task.audio_file_path_absolute)

        with open(aeneas_task.sync_map_file_path_absolute, "r") as file:
            sync_map = json.load(file)

        output_directory = f"{res_folder_path}/{user_id}"
        for id, fragment in enumerate(sync_map['fragments']):
            start_time = float(fragment['begin']) * 1_000
            end_time = float(fragment['end']) * 1_000
            segment = audio[start_time:end_time]

            segment_name = f"{index}_{id}.wav"
            segment.export(f"{output_directory}/{segment_name}", format="wav")

            res.append({
                'index': f"{index}_{id}",
                'user_id': user_id,
                'description': fragment['lines'][0],
            })

pd.DataFrame(res).to_csv(f"{res_folder_path}/users.csv", index=False)
```
